```python
import requests
from bs4 import BeautifulSoup 
from tqdm import tqdm
import pandas as pd
```

```python
df2 = pd.read_csv("1000_book_data.csv")
```

```python
df2.head()
```

```python
data = []
for link in tqdm(df2["book_link"]):
    res = requests.get(link)
    soup = BeautifulSoup(res.text , "html.parser")
    
    book_type = soup.find("ul" , class_ = "breadcrumb").find_all("a")[2].text
    
    temp = soup.find("table" , class_ = "table table-striped").find_all("td")
    upc                  = temp[0].text
    product_type         =temp[1].text
    price_excl_tax       = temp[2].text[2:]
    price_incl_tax       = temp[3].text[2:]
    tax                  = temp[4].text[2:]
    availability         = temp[5].text
    number_of_reviews    = temp[6].text
    
    data.append([book_type,upc,product_type,price_excl_tax,price_incl_tax,tax,availability,number_of_reviews])
    

    
    
```

```python
df2 = pd.DataFrame(data , columns = ["booktype" , "upc", "product_type", "price_excl_tax", "price_incl_tax" ,"tax", "availability" ,"number_of_reviews"])
```

```python
df2.to_csv("1000_books_data_2.csv")
```

```python
final_data = pd.DataFrame()

final_data["book_title"] = df1["book_title"]
final_data["book_type"] = df2["booktype"]
final_data["book_price"] = df1["price"]
final_data["book_availability"] = df2["availability"]
final_data["product_type"] = df2["product_type"]
final_data["book_price_excl_tax"] = df2["price_excl_tax"]
final_data["book_price_incl_tax"] = df2["price_incl_tax"]
final_data["book_tax"] = df2["tax"]
final_data["book_no_of_reviews"] = df2["number_of_reviews"]
final_data["book_stock"] = df1["stock"]
final_data["book_link"] = df1["book_link"]
final_data["book_img_link"] = df1["img_link"]
```
